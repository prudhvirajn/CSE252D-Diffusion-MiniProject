## **Conclusion**

Given some text prompt key points in a story, we can generate images for each prompt and then interpolate between the key points to easily create some frames between actions. Although interpolation steps do a good job of maintaining the semantics of the actions between frames, it is difficult to keep overall consistency throughout the story because each story key point is generated individually. This can be seen through the visual change of people and background despite the overall semantic action remaining consistent. We attempted to tackle this problem through style transfer methods, but current works struggle to keep objects, backgrounds, and people consistent throughout. We also tried being much more descriptive in our prompts in order to lead to less randomness between story key points, however it is still not perfect. To improve this, further work can be done on improving object and background consistency between story key points.
